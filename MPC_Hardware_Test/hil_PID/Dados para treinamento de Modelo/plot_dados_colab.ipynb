```python
import pandas as pd
import matplotlib.pyplot as plt

# No Google Colab, se você fizer o upload direto, os arquivos ficarão na pasta raiz '/content'
pasta_dados = '.'
```

## Análise: `sinal1_semi_estatica_malha_fechada.csv`


```python
arquivo = f"{pasta_dados}/sinal1_semi_estatica_malha_fechada.csv"
df = pd.read_csv(arquivo)

plt.figure(figsize=(12, 5))
if 'Tempo_s' in df.columns:
    if 'Referencia' in df.columns:
        plt.plot(df['Tempo_s'], df['Referencia'], label='Referência', linestyle='--', color='black')
    if 'Saida_y' in df.columns:
        plt.plot(df['Tempo_s'], df['Saida_y'], label='Saída (y)', color='blue')
    if 'Controle_u' in df.columns:
        plt.plot(df['Tempo_s'], df['Controle_u'], label='Controle (u)', color='red', alpha=0.7)
    plt.xlabel('Tempo (s)')
else:
    for col in df.columns[1:]:
        plt.plot(df[df.columns[0]], df[col], label=col)
    plt.xlabel(df.columns[0])

plt.ylabel('Amplitude')
plt.title('sinal1_semi_estatica_malha_fechada.csv')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
